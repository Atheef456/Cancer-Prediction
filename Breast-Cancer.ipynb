```python
import pandas as pd

df = pd.read_csv("breast-cancer-dataset.csv")  # Replace with your file name
```

Using MATPLOT Library to showing the graphs..........

```python
import matplotlib.pyplot as plt

plt.hist(df["Tumor Size (cm)"],bins=20, color='red', edgecolor='black')
plt.xlabel("Tumor Size")
plt.ylabel("Frequency")
plt.title("Distribution of Tumor Sizes")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=df["Diagnosis Result"], y=df["Age"],hue=df["Diagnosis Result"],palette="coolwarm")
plt.xlabel("Diagnosis (Benign/Malignant)")
plt.ylabel("Age")
plt.title("Age vs Diagnosis")
plt.show()
```

## **Age vs Tumor Size**
Find the relation between the coulmns ***Age***, ***Tumor Size*** and ***Diagnosis result***

```python
sns.relplot(x=df["Age"], y=df["Tumor Size (cm)"], hue=df["Diagnosis Result"],col=df["Diagnosis Result"],palette="coolwarm")
plt.xlabel("Age")
plt.ylabel("Tumor Size")
plt.show()
```

```python
import numpy as np

# Select only numeric columns
numeric_df = df[['Menopause', 'Metastasis']];
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
